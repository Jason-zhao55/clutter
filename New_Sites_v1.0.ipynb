```python
Gsm_Lq=Gsm_Lq.dropna()
Gsm_Lq[[ '本地小区标识', 'eNodeB标识', '下行频点', '频带', 'CI','LAC','TRX1/BCCH','NCC','BCC']]=Gsm_Lq[[ '本地小区标识', 'eNodeB标识', '下行频点', '频带', 'CI','LAC','TRX1/BCCH','NCC','BCC']].astype(int)
# 定义向G网的xlow参数
def xlow_gsm(a_g):
    if a_g==38950 or a_g==39148 or a_g==39292:
        return 25
    else:
        return 10
Gsm_Lq['THRESHXLOW']=Gsm_Lq.apply(lambda x : xlow_gsm(x['下行频点']),axis=1)
Gsm_Lq=Gsm_Lq[['LTE小区名', 'GSM邻区小区名', 'NAME', '本地小区标识', 'eNodeB标识', '下行频点', '频带', 'THRESHXLOW', 'CI',
       'LAC', 'TRX1/BCCH', 'NCC', 'BCC']]

# 定义创建GERAN相邻频点组命令
def g_fre_grouped(a1_gsm,b1_gsm,c1_gsm):
    return f'''ADD GERANNFREQGROUP:LOCALCELLID={a1_gsm},BCCHGROUPID=0,GERANVERSION=EDGE,STARTINGARFCN=0,BANDINDICATOR=GSM_dcs1800,CELLRESELPRIORITYCFGIND=CFG,CELLRESELPRIORITY=3,PMAXGERANCFGIND=NOT_CFG,THRESHXLOW={b1_gsm},CONNFREQPRIORITY=8;{{{c1_gsm}}}'''
# 定义创建GERAN BCCH相邻频点命令
def g_fre_bcch(d1_gsm,e1_gsm,f1_gsm):
    return f'''ADD GERANNFREQGROUPARFCN:LOCALCELLID={d1_gsm},BCCHGROUPID=0,GERANARFCN={e1_gsm};{{{f1_gsm}}}'''
# 定义创建GERAN 外部
def g_wb(h1_gsm,i1_gsm,j1_gsm,k1_gsm,l1_gsm,m1_gsm,n1_gsm):
    return f'''ADD GERANEXTERNALCELL:MCC="460",MNC="00",GERANCELLID={h1_gsm},LAC={i1_gsm},RACCFGIND=NOT_CFG,BANDINDICATOR=GSM_dcs1800,GERANARFCN={j1_gsm},NETWORKCOLOURCODE={k1_gsm},BASESTATIONCOLOURCODE={l1_gsm},CELLNAME="{m1_gsm}";{{{n1_gsm}}}'''

# 定义创建GERAN邻区关系
def g_lq(p1_gsm,q1_gsm,r1_gsm,s1_gsm,t1_gsm,u1_gsm):
    return f'''ADD GERANNCELL:LOCALCELLID={p1_gsm},MCC="460",MNC="00",LAC={q1_gsm},GERANCELLID={r1_gsm},BLINDHOPRIORITY=2,LOCALCELLNAME="{s1_gsm}",NEIGHBOURCELLNAME="{t1_gsm}";{{{u1_gsm}}}'''

# 定义运营商共享信息
def share(v1_gsm,w1_gsm):
    return f'''ADD GERANRANSHARE:LOCALCELLID={v1_gsm},BCCHGROUPID=0,MCC="460",MNC="00",CELLRESELPRIORITYCFGIND=NOT_CFG;{{{w1_gsm}}}'''
# 定义创建Geran小区重选
def cellresel(a2_gsm,b2_gsm):
    return f'''ADD CELLRESELGERAN:LOCALCELLID={a2_gsm},SPEEDSTATESFCFGIND=NOT_CFG;{{{b2_gsm}}}'''


Gsm_Lq['mml']=Gsm_Lq.apply(lambda x : g_fre_grouped(x['本地小区标识'],x['THRESHXLOW'],x['NAME']),axis=1)
Gsm_Lq['mml1']=Gsm_Lq.apply(lambda x : g_fre_bcch(x['本地小区标识'],x['TRX1/BCCH'],x['NAME']),axis=1)
Gsm_Lq['mml2']=Gsm_Lq.apply(lambda x : g_wb(x['CI'],x['LAC'],x['TRX1/BCCH'],x['NCC'],x['BCC'],x['GSM邻区小区名'],x['NAME']),axis=1)
Gsm_Lq['mml3']=Gsm_Lq.apply(lambda x : g_lq(x['本地小区标识'],x['LAC'],x['CI'],x['LTE小区名'],x['GSM邻区小区名'],x['NAME']),axis=1)
Gsm_Lq['mml4']=Gsm_Lq.apply(lambda x : share(x['本地小区标识'],x['NAME']),axis=1)
Gsm_Lq['mml5']=Gsm_Lq.apply(lambda x : cellresel(x['本地小区标识'],x['NAME']),axis=1)
tjlq_gsm=Gsm_Lq[['mml','mml1','mml2','mml3','mml4','mml5']]
tjlq_gsm=tjlq_gsm.unstack(level=-1)
tjlq_gsm=pd.DataFrame(tjlq_gsm)
tjlq_gsm.rename(columns={0:'mml'},inplace=True)
tjlq_gsm.drop_duplicates(inplace=True)
tjlq_gsm.dropna(subset=['mml'],inplace=True)

# 删除空值
while None in tjlq_gsm:
    tjlq_gsm.remove(None)
tjlq_gsm.to_csv('3_TJGSMLQ.txt',index=False,header=False,encoding='gbk',sep='?',quoting=3,doublequote=False,escapechar='?')
```
